```python
import pandas as pd
import numpy as np
from pandas import DataFrame

from sklearn import tree
import sklearn
import matplotlib.pyplot as plt # using visualization package
# You can also export the tree in Graphviz format using the export_graphviz exporter
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.model_selection import train_test_split # Import train_test_split function
all_data = pd.read_csv("/Users/clause/documents/GitHub/bank_data_analysis/bank-full.csv",sep=";")
```

```python
for i in range(len(all_data)):
    if all_data["month"][i]=="jan":
        all_data["month"][i]=1
    elif all_data["month"][i]=="feb":
        all_data["month"][i]=2
    elif all_data["month"][i]=="mar":
        all_data["month"][i]=3
    elif all_data["month"][i]=="apr":
        all_data["month"][i]=4
    elif all_data["month"][i]=="may":
        all_data["month"][i]=5
    elif all_data["month"][i]=="jun":
        all_data["month"][i]=6
    elif all_data["month"][i]=="jul":
        all_data["month"][i]=7
    elif all_data["month"][i]=="aug":
        all_data["month"][i]=8
    elif all_data["month"][i]=="sep":
        all_data["month"][i]=9
    elif all_data["month"][i]=="oct":
        all_data["month"][i]=10
    elif all_data["month"][i]=="nov":
        all_data["month"][i]=11
    elif all_data["month"][i]=="dec":
        all_data["month"][i]=12
```

```python
feature_cols = ['age','job','marital','education','default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome']
all_data=all_data.replace("yes",1)
all_data=all_data.replace("no",0)
```

```python
X=all_data[feature_cols]
y=all_data["y"]
```

```python
from sklearn import preprocessing
import numpy as np
```

```python
X=X.replace("yes",1)
X=X.replace("no",0)
```

```python
numerical_cols=['age','balance','campaign','pdays','previous','duration',"default","housing","loan"]
cats_cols=['job','marital','education','contact','poutcome']
```

```python
X_cats=X[cats_cols]
```

```python
X_cats=np.array(X_cats)
from sklearn.preprocessing import OneHotEncoder
enc = preprocessing.OneHotEncoder()
enc.fit(X_cats)    # fit来学习编码
X_cats=enc.transform(X_cats).toarray()    # 进行编码
X_cats=pd.DataFrame(X_cats)
```

```python
X_numerical=X[numerical_cols]
```

```python
X_time=X[["day","month"]]
X_alltime=30*X["month"]+X["day"]
X_alltime=X_alltime/360
```

```python
X=pd.concat([X_alltime,X_cats,X_numerical],axis=1)
```

```python
colnames=["time","onehot1","onehot2","onehot3","onehot4","onehot5","onehot6","onehot7","onehot8","onehot9","onehot10","onehot11","onehot12",
          "onehot13","onehot14","onehot15","onehot16","onehot17","onehot18","onehot19","onehot20","onehot21","onehot22","onehot23",
         "onehot24","onehot25","onehot26",'age','balance','campaign','pdays','previous','duration',"default","housing","loan"]
cats_cols=["onehot1","onehot2","onehot3","onehot4","onehot5","onehot6","onehot7","onehot8","onehot9","onehot10","onehot11","onehot12",
          "onehot13","onehot14","onehot15","onehot16","onehot17","onehot18","onehot19","onehot20","onehot21","onehot22","onehot23",
         "onehot24","onehot25","onehot26"]
numerical_cols=["time",'age','balance','campaign','pdays','previous','duration']
binary_cols=["default","housing","loan"]
```

```python
X.columns=colnames
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python
train_data=pd.concat([X_train, y_train], axis=1)
```

```python
train_data_1=train_data[train_data["y"]==1]
train_data_0=train_data[train_data["y"]==0]
```

```python
train_data_1=train_data_1.sample(5000, replace=True)
train_data_0=train_data_0.sample(5000, replace=True)
```

```python
train_data = pd.concat([train_data_0, train_data_1], axis=0)
```

```python
train_data
```

```python
balance_mean=sum(train_data["balance"])/len(train_data["balance"])
balance_std=np.std(train_data["balance"])

age_mean=sum(train_data["age"])/len(train_data["age"])
age_std=np.std(train_data["age"])

pdays_mean=sum(train_data["pdays"])/len(train_data["balance"])
pdays_std=np.std(train_data["pdays"])

duration_mean=sum(train_data["duration"])/len(train_data["duration"])
duration_std=np.std(train_data["duration"])


previous_mean=sum(train_data["previous"])/len(train_data["duration"])
previous_std=np.std(train_data["previous"])

campaign_mean=sum(train_data["campaign"])/len(train_data["duration"])
campaign_std=np.std(train_data["campaign"])


train_data["balance"]=(train_data["balance"]-balance_mean)/balance_std
X_test["balance"]=(X_test["balance"]-balance_mean)/balance_std

train_data["age"]=(train_data["age"]-age_mean)/age_std
X_test["age"]=(X_test["age"]-age_mean)/age_std


train_data["pdays"]=(train_data["pdays"]-pdays_mean)/pdays_std
X_test["pdays"]=(X_test["pdays"]-pdays_mean)/pdays_std

train_data["duration"]=(train_data["duration"]-duration_mean)/duration_std
X_test["duration"]=(X_test["duration"]-duration_mean)/duration_std

train_data["previous"]=(train_data["previous"]-previous_mean)/previous_std
X_test["previous"]=(X_test["previous"]-previous_mean)/previous_std

train_data["campaign"]=(train_data["campaign"]-campaign_mean)/campaign_std
X_test["campaign"]=(X_test["campaign"]-campaign_mean)/campaign_std
```

```python
train_data.isna().sum()
```

```python
from sklearn.svm import LinearSVC
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectFromModel
```

```python
lsvc = LinearSVC(C=0.01, penalty="l1", dual=False).fit(X_train, y_train)
model = SelectFromModel(lsvc, prefit=True)
X_train = model.transform(X_train)
```

```python
X_test=model.transform(X_test)
```
